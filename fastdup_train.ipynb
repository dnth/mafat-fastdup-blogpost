```python
import fastdup
```

```python
fastdup.run(input_dir="./dataset/train/images_jpg", 
            work_dir="./fastdup_report/train")
```

```python
from IPython.display import HTML
fastdup.create_duplicates_gallery(similarity_file='./fastdup_report/train/similarity.csv',
                                  save_path='./fastdup_report/train/', 
                                  num_images=20)

HTML('./fastdup_report/train/similarity.html')
```

```python
# Find darkest images
fastdup.create_stats_gallery('./fastdup_report/train/atrain_stats.csv', save_path='./fastdup_report/train',max_width=400,metric='mean',num_images=20)
HTML('./mean.html')
```

```python
# Find brightest images
fastdup.create_stats_gallery('./fastdup_report/train/atrain_stats.csv',save_path='./fastdup_report/train',max_width=400,metric='mean', descending=True,num_images=20)
HTML('./mean.html')
```

```python
fastdup.create_stats_gallery('./fastdup_report/train/atrain_stats.csv', save_path='./fastdup_report/train',max_width=400,metric='stdv',descending=False, num_images=5)
HTML('./stdv.html')
```

```python
fastdup.create_outliers_gallery(outliers_file='./fastdup_report/train/outliers.csv',            
                                save_path='./fastdup_report/train/', 
                                num_images=50)
HTML('./fastdup_report/train/outliers.html')
```
